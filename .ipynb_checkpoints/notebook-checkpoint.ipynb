## Text preprocessing

### Capitalization

```python
data["message"] = data["message"].str.lower()
```

### Tokenization

```python
data["tokens"] = data["message"].apply(lambda x : nltk.word_tokenize(x))
```

### Removing special characters

```python
data["tokens"] = data["tokens"].apply(lambda x : list(filter(str.isalpha,x)))
```

### Removing stop words

```python
stopwords = set(nltk.corpus.stopwords.words("english"))
```

```python
data["tokens"] = data["tokens"].apply(lambda x : [w for w in x if w not in stopwords])
```

### Stemming

```python
stemmer = nltk.stem.porter.PorterStemmer()
```

```python
data["tokens"] = data["tokens"].apply(lambda x : list(map(stemmer.stem, x)))
```

## Modeling,training and feature extraction

### Feature extraction

```python
count_vextorizer = CountVectorizer()
```

```python
count_vextorizer.fit(data["transformed_text"])
```

## Evaluation

```python
metrics = [accuracy_score,f1_score,confusion_matrix,precision_score,recall_score]
models  = [gnb,tree_classifier,sgd_classifier,logistic_regressor,k_neighbors_classifier,svc,random_forest_classifier]
```

```python
def evaluate(models, metrics, X, y):

    algorithms = [model["estimator"].__class__.__name__ for model in models]
    columns = ["Algorithm"] + [metric.__name__ for metric in metrics]
    results = np.zeros(shape=(len(models),len(columns))).tolist()
    
    for i,model in enumerate(models):
        
        y_hat = model.predict(X)
        results[i][0] = algorithms[i]
        
        for j,metric in enumerate(metrics):
            
            value = metric(y, y_hat)
            results[i][j+1] = value

    return pd.DataFrame(data=results,columns=columns)
```

```python
evaluation_df = evaluate(models, metrics, X_test, y_test)
evaluation_df
```

```python
melted_df = pd.melt(evaluation_df, id_vars=['Algorithm'], var_name='Metric', value_name='Value')
melted_df.drop(index=melted_df.index[melted_df["Metric"] == "confusion_matrix"], inplace=True)
```

```python
sns.barplot(x='Algorithm', y='Value', hue='Metric', data=melted_df)
plt.xticks(rotation=80)
```

- From the results above the best model is LogisticRegression.

### Building a Meta Classifier

#### Boosting

```python
voting_classifier = VotingClassifier(estimators=[
    ("sgd_classifier", sgd_classifier),
    ("logistic_regressor", logistic_regressor),
    ("random_forest_classifier",random_forest_classifier)
], voting = "soft")
```

```python
voting_classifier = Pipeline(steps=[
    ("estimator", voting_classifier)
])
```

```python
voting_classifier.fit(X_train, y_train)
```

### Stacking

```python
staking_classifier = StackingClassifier(estimators=[
    ("sgd_classifier", sgd_classifier),
    ("logistic_regressor", logistic_regressor),
    ("random_forest_classifier",random_forest_classifier)
], final_estimator = RandomForestClassifier())
```

```python
staking_classifier = Pipeline(steps=[
    ("estimator", staking_classifier)
])
```

```python
staking_classifier.fit(X_train,y_train)
```

```python
evaluation_df = evaluate([voting_classifier,staking_classifier], metrics, X_test, y_test)
evaluation_df
```

- VotingClassifier is a liite bit better than LogesticRegression model.

## Export the best model

#### Build a pipeline containing all the preprocessing

```python
def transform_text(text):
    text = text.lower()
    tokens = nltk.word_tokenize(text)
    tokens = [w for w in tokens if str.isalpha(w)]
    tokens = [w for w in tokens if w not in stopwords]
    tokens = [stemmer.stem(w) for w in tokens]
    return ' '.join(tokens)
```

```python
def transform_ds(X):
    return np.array(list(map(transform_text, X)))
```

```python
def length(X):
    return np.array([[len(x)] for x in X])
```

```python
def densify(X):
    return X.toarray()
```

```python
text_preprocessor = Pipeline(steps=[
    ("transforme_text",FunctionTransformer(func=transform_ds)),
    ("count_vextorizer",count_vextorizer),
    ("densify",FunctionTransformer(func=densify)),
    ("reduce_dimension",pca)
])
```

```python
features_preprocessor = FeatureUnion(transformer_list=[
    ("length", FunctionTransformer(func=length)),
    ("text", text_preprocessor),
])
```

```python
model = Pipeline(steps=[
    ("feature",features_preprocessor),
    ("estimator",logistic_regressor)
])
```
